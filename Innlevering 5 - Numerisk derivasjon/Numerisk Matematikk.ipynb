## (a)

#### Hva er en sekant?

En sekant er en linje som skjærer en graf i minst to punkter.

#### Forklar hvorfor stigningstallet $k$ til tangenten gjennom punktet a på funksjonen $f$ kan skrives slik: 
$$k = \frac{f(x + \Delta x) - f(x)}{\Delta x}$$

Stigningstallet vil kunne skrives som gjennomsnittlig verdiendringen, og kan dermed skrives på måten ovenfor.

## (b)

#### Hva er en tangent?

En tangent er stigningstallet til et området tilnærmet et punkt.

#### Forklar hvorfor stigningstallet $k$ til tangenten gjennom punktet $a$ på funksjonen $f$ kan skrives slik:
$$f'(a)= \lim_{{\Delta a \to 0}}\frac{f(x + \Delta x) - f(x)}{\Delta x}$$

Stigningstallet vil kunne skrives som gjennomsnittlig verdiendring over en tid tilnærmet null. Da får vi stigningtallet til et tilnærmet punkt.

## (c)

#### Forklar nå med ord hva derivasjon er.

I matematikk er derivasjon en operasjon hvor man bestemmer den deriverte av en funksjon.

## (d)

#### Newtons kvotient kan formuleres slik:
$$f'(x) = \frac{f(x+h)-f(x)}{h}$$

#### Hva er forskjellen mellom Newtons kvotient og definisjonen av den deriverte? Implementer metoden som en funksjon i Python og test den på funksjoner du klarer å derivere analytisk.

Forskjellen mellom Newtons kvotient og definisjonen av den deriverte er at Newtons kvotient tar ibruk en numerisk metode og setter forskjellen arbitrært lite.

# Eksempler på derivasjon numerisk og av diskrete data

```python
from pylab import *

def deriver(x, h = 1E-8):
    return (f(x + h) - f(x)) / h

def f(x):
    return 2 * x ** 2 + 5 * x

x = [x for x in range(-4, 5)]
y = [f(x) for x in range(len(x))]

plot(x,y)

y = [deriver(x) for x in range(len(x))]

plot(x,y)
show()
```

```python
from pylab import *

data = genfromtxt('Pressureandvelocity.csv', delimiter=',', skip_header=3)

x = [data[x][0] for x in range(len(data))]
y = [data[x][2] for x in range(len(data))]

xlabel("Tid (s)")
ylabel("Høyde (m)")

plot(x,y)
show()


def deriver(x):
    try:
        return (data[x + 1][2] - data[x][2]) / (data[x + 1][0] - data[x][0])
    except ZeroDivisionError:
        return None

x.pop(-1)
y = [deriver(x) for x in range(len(data) - 1)]

xlabel("Tid (s)")
ylabel("Fart (m/s)")

plot(x,y)
show()
```
