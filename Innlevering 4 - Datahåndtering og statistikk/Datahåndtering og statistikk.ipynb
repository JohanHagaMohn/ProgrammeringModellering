# Statistisk analyse

## Gjennomsnitt

Gjennomsnitt er definert som summen av alle elementer delt på antall elementer.



## Standardavvik

Standardavvik er definert som gjennomsnittsavstand fra gjennomsnittsverdien.


```python
from pylab import *

data = loadtxt("sunspots.csv", delimiter=",", skiprows=1)

x = data[:, 0]
y = data[:, 1]

gjennomsnitt = mean(y)
standardavvik = std(y)

print(f"Gjennomsnittet er {gjennomsnitt} og standardavviket er {standardavvik}!")
```

# Regresjon

Regresjon er å finne et generelt utrykk ut ifra enkeltmålinger.

## Plotting

For å gjøre det, benytter vi oss av pylabs polyfit-funksjon for å få en polynomfunksjon vi kan plotte.

```python
from pylab import *

data = array(loadtxt("sunspots.csv", delimiter=",", skiprows=1))

x = data[:, 0]
y = data[:, 1]

grad = int(input("Grad: "))
reg = polyfit(x, y, grad)
yreg = [reg[- 1] for i in range(len(x))]

for i in range(grad):
    yreg += reg[- 2 - i] * x ** (i + 1)

scatter(x, y)
plot(x, yreg)
show()
```

## Analyse

For å analysere resultatet, kan vi finne determinasjonskoeffisienten for å finne ut hvor godt tilpasset funksjonen vår er.

r2_score-funksjonen tar y-verdier fra to grafer som input, og gir oss et tall mellom 0 og 1 for å symbolisere dette.


```python
from pylab import *
from sklearn.metrics import r2_score

data = array(loadtxt("sunspots.csv", delimiter=",", skiprows=1))

x = data[:, 0]
y = data[:, 1]

grad = int(input("Grad: "))
reg = polyfit(x, y, grad)
yreg = [reg[- 1] for i in range(len(x))]

for i in range(grad):
    yreg += reg[- 2 - i] * x ** (i + 1)

score = r2_score(y, yreg)

print(score)
```

## Konklusjon

Eksemplet over har en r2-score mellom 0.1 og 0.2 utifra graden til polynomfunksjonen. Dette sier oss at en slik polynomfunksjon kan være ganske dårlig til å modellere solflekker.

# Datafremstilling

## Kakediagram

Vi kan bruke pie-funksjonen i pylab til å lage kakediagrammer.


```python
from pylab import *

fag_med_andreas_haraldsrud_som_lærer = ["Naturfag", "Promod", "Kjemi"]
elever = [50, 30, 40]

pie(elever, labels=fag_med_andreas_haraldsrud_som_lærer)
show()
```

```python
from pylab import *

fag = {
    "Naturfag": 50,
    "Kjemi": 40,
    "Promod": 30
}

pie(fag.values(), labels=fag, explode=[0,0,3], colors=["yellow", "yellow", "black"], startangle=45)
title("Ganske brukbare fag noen ganger")
show()
```

## Søylediagram

Vi kan bruke bar-funksjonen i pylab til å lage søylediagrammer.


```python
from pylab import *

x = [x for x in range(0, 10000, 1000)]
y = ["{:.0f}M".format(i**2*10E-6) for i in x]

farger = ["white", "crimson", "red", "darkorange", "orange", "yellow", "limegreen", "seagreen","green", "darkgreen"]

bar(x, y, width=900, yerr=[x*10E-5 for x in x], color=farger)
title("Nyttighet av realfag")
xlabel("Timer med realfag")
ylabel("Fremtidig inntekt (usd)")
show()
```
